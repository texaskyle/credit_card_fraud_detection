# Handling Outliers

In my provided dataset, the data has already being transformed using Using PCA, therefore i can make an assumption that outliers had been already handled. since PCA comes later after outliers detection.

```python
data_fraud = data[data['Class'] == 1]
not_data_fraud = data[data['Class'] == 0]
```

# Distribution of classes with respect to time

```python
plt.figure(figsize=(8,5))
ax = sns.distplot(data_fraud['Time'], label='fraudulent', hist=False)
ax = sns.distplot(not_data_fraud['Time'], label='non fraudulent', hist=False)
ax.set(xlabel='Seconds elapsed between the transction and the first transction')
plt.legend()
plt.grid()
plt.show()
```

There is no specific pattern of fraudlent and non fraudlent with respect to time, hence the column time can be drop.

```python
data = data.drop("Time", axis=1)
data
```

# Distribution of classes with respect to amount

```python
plt.figure(figsize=(6,4))
ax = sns.distplot(data_fraud['Amount'], label='fraudulent', hist=False)
ax.set(xlabel='Transction Amount')
plt.legend()
plt.show()
```

I can make an implication that fraudulent cases are most frequent for a low transaction amount from 0 to 2000

```python
plt.figure(figsize=(6,4))
ax = sns.distplot(not_data_fraud['Amount'], label='not fraudulent', color='orange', hist=False)
ax.set(xlabel='Transction Amount')
plt.legend()
plt.show()
```

Non fraudulent cases are spend from 0 to 25000
